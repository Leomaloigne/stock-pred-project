```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import yfinance as yf
from keras.models import Sequential
from keras.layers import Dense, LSTM

ticker_select = 'AAPL'

df = yf.download(ticker_select, start= '2017-01-01', end = '2024-01-01')

print (df.head())

print (df.shape)

ticker = yf.Ticker(ticker_select)
info = ticker.info
print (info)

data = df['Close']

print (data.head())

data.plot(title ='closing price')
plt.xlabel('date')
plt.ylabel('closing price (USD)')
plt.show()

dataset = data.values
print (len(dataset))
training_data_len = math.ceil(len(dataset) * 0.8)
print (training_data_len)


scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data.values.reshape(-1, 1))
print (scaled_data)

train_data = scaled_data[0:training_data_len, :]
x_train = []
y_train = []

days = 60


for i in range (days, len(train_data)):
    x_train.append(train_data[i-days:i, 0])
    y_train.append(train_data[i, 0])
    if i<= days + 2:
        print(x_train)
        print(y_train)
        print()

x_train, y_train = np.array(x_train), np.array(y_train)

x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
print (x_train.shape)

model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
model.compile(optimizer = 'adam', loss ='mean_squared_error')

model.fit(x_train, y_train, batch_size=32, epochs=10)

test_data = scaled_data[training_data_len-days:, :]

x_test = []
y_test = dataset[training_data_len:,:]

for i in range(days, len(test_data)):
    x_test.append(test_data[i-days:i, 0])
    

x_test = np.array(x_test)

x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

preds = model.predict(x_test)
preds = scaler.inverse_transform(preds)


rmse = np.sqrt(np.mean((preds - y_test)**2))
print (rmse)

train = df[:training_data_len]
tests = df[training_data_len:]
tests['preds'] = preds


plt.title('model')
plt.plot(train['Close'])
plt.plot(tests[['Close', 'preds']])
plt.legend(['train', 'test','preds'])

last_x_days = data[-days:].values
last_x_days_scaled = scaler.transform(last_x_days)
x_day_test = []
x_day_test.append(last_x_days_scaled)
x_day_test =np.array(x_day_test)
x_day_test = np.reshape(x_day_test, (x_day_test.shape[0], x_day_test.shape[1], 1))

pred_price = model.predict(x_day_test)
pred_price = scaler.inverse_transform(pred_price)
print (pred_price)
```
